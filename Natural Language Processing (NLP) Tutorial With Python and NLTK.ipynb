```python
from nltk.tokenize import word_tokenize
```

```python
quote = "Taking responsibility - practicing 100 percent responsibility every day - is about seeing ourselves not as right or wrong, but as an agent, chooser, problem solver, and learner in the complex interrelationships of our lives so that we can better integrate with the people and world around us. When we do this, we enjoy a better and more productive way to live and lead."
```

```python
# provides the data type of our quote. It is important to note that when using NLP packages str is the prefered data type.
type(quote)
```

```python
# converts out quote into tokens or indidual pieces. 
quote_Tokens = word_tokenize(quote)
quote_Tokens
```

```python
# provides the length of tokens extracted from our quote
len(quote_Tokens)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
# converts the string to lower case then counts it to make sure the same word isnt counted differently if its upper case. 
for word in quote_Tokens:
    fdist[word.lower()]+=1
fdist
```

```python
fdist['quote']
```

```python
# provides the length of our frequency distribution
len(fdist)
```

```python
# provides the top 10 most recurring values
fdist_top10 = fdist.most_common(10)
fdist_top10
```

```python
# Searches for linebreaks and breaks up the string into paragraphs that you can call
from nltk.tokenize import blankline_tokenize
quote_Blank = blankline_tokenize(quote)
len(quote_Blank)
```

```python
# Prints paragraph 1 (Python starts at 0 for counting so len is 1 but to call you use [0])
quote_Blank[0]
```

## Tokenization Cont. 
### Tokens are separated into Bigrams - tokens containing two consecutive words Trigrams - Tokens containing 3 consecutive words or NGrams - Tokens containing any number of consecutive written words

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
# Tokenization of our string and a list is then printed. 
string = "The best and most beautiful things in the world cannot be seen or even touched, they must be felt with the heart."
quotes_tokens = nltk.word_tokenize(string)
quotes_tokens
```

```python
#Provides a list of bigrams
quotes_bigrams = list(nltk.bigrams(quotes_tokens))
quotes_bigrams
```

```python
#Provides a list of trigrams
quotes_trigrams = list(nltk.trigrams(quotes_tokens))
quotes_trigrams
```

```python
# provides a list of pentagrams (n_grams with 5 selected as n)
quotes_ngrams = list(nltk.ngrams(quotes_tokens, 5))
quotes_ngrams
```

# STEMMING
## Stemming normalizes words into their base or root forms. I.E. Justified --> Justice
## Affectation Affects Affections Affected Affection Affecting all become Affect

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
# You can call a single word to obtain the root form quickly
pst.stem("having")
```

```python
# here we will generate a list and have teh PorterStemmer package stem for us
words_to_stem = ["give","giving", "given", "gave"]
for words in words_to_stem:
    print(words+ ":" +pst.stem(words))
```

```python
# Here we use LancasterStemmer which is a much more aggressive stemmer giving use the root three letters in this instance
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
for words in words_to_stem:
    print(words+ ":" +lst.stem(words))
```

```python
# it is important to note that SnowballStemmer requires the language be declared \
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')
```

```python
#Snowball yields the same results for us in this instance but does use a different stemming method
for words in words_to_stem:
    print(words+ ":" +sbst.stem(words))
```

# PARTS OF SPEECH (POS)


## POS Example:


```python
#Declaring a new string and then tokenizing the string
sent = "Timothy is a natural when it comes to drawing."
sent_tokens = word_tokenize(sent)
```

```python
# after tokenizing we get the Part of Speech tags using this code.
for token in sent_tokens:
    print(nltk.pos_tag([token]))
```

```python
# Tokenizing and tagging a new string
sent2 = "John is eating a delicious cake."
sent2_tokens = word_tokenize(sent2)
for token in sent2_tokens:
    print(nltk.pos_tag([token]))
```

# Named Entity Recognition
## Movie Names <br> Monetary Value(Dollar) <br> Organization (Wal-Mart) <br> Location (New York) <br> Quantities (40) <br> Person (Larry)
## 3 Phases of NER <br> Noun Phase (Extracts Nouns and POS Tages) <br> Phrase Classification (extracted nouns are classified into catagories) <br> Entity Disambiguation - Reclassifying incorrectly classified nouns. 
## Example:
### Google's CEO Sundar Pichai introduced the new Pixel at Minnesota Roi Centre Event.
### Google is an Organization. Sundar Pichai is a person. Minnesota is a location. Roi Centre Event is an Organization. 
### This is an additional layer on top of the POS Tagging

```python
from nltk import ne_chunk
```

```python
NE_sent = "The US President stays in the White House."
```

```python
# First we tokenize the string and then we tag it
NE_tokens = word_tokenize(NE_sent)
NE_tag = nltk.pos_tag(NE_tokens)
```

```python
# Classifies and declares the Named Entities
NE_NER = ne_chunk(NE_tag)
print(NE_NER)
```

# CHUNKING
## Picking up individual pieces of information adn grouping them into bigger pieces
## Bigger pieces are chunks

```python
# a new string that we are tokenizing and pos_tagging to prep it for chunking
new = "The big cat ate the little mouse who was after the fresh cheese."
new_tag = nltk.pos_tag(word_tokenize(new))
new_tag
```

```python
# Uses Regex to delcare the grammar structure
grammar_np = r"NP: {<DT>?<JJ>*<NN>}"
```

```python
chunk_parser = nltk.RegexpParser(grammar_np)
```

```python
# Generates the Syntax Tree using Ghostscript
chunk_result = chunk_parser.parse(new_tag)
chunk_result
```
